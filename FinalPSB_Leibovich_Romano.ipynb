## Detección de complejos QRS (canal V5)

Para el detector de complejos QRS se utiliza el algoritmo de Pan y Tompkins. Para esto se construye la siguiente función, que puede o no mostrar los gráficos de todos los pasos.

```python
def detector_QRS(senal, fs, segundos=5, plotting = True):
  x=senal
  L = len(x)
  t = np.linspace(0,L/fs,L)

  #FILTRO PASABANDAS
  wn = [5,15]
  orden_N = 3
  [b, a] = scipy.signal.butter(orden_N, wn, btype='bandpass', fs=fs)
  xfiltrada = scipy.signal.filtfilt(b,a,x)
  
  #DERIVACION
  paso = 2 
  skip_factor = 2*paso
  ts = 1/fs 
  x_deriv = []

  for i in range(0,L-2*paso):
    x_deriv.append((xfiltrada[i+paso] - xfiltrada[i-paso])/(2*paso*ts))

  ceros = np.zeros(2*paso)
  x_deriv.extend(ceros)

  #CUADRADO PARA POSITIVIZAR
  x_cuad = []
  for i in x_deriv:
    x_cuad.append(i**2)

  #INTEGRADOR
  muestras = int(0.15*fs)
  x_integ = np.zeros(L)

  for i in range(0,L-muestras):
    x_integ[i] = (np.sum(x_cuad[i:i+muestras])/muestras)

  ceros = np.zeros(muestras)
  np.append(x_integ,ceros)

  #PICOS
  umbral = 0.25*max(x_integ)
  picos, _ = scipy.signal.find_peaks(x_integ,distance=0.45*fs, height=umbral)

  #CORRECION RETRASOS
  retraso = int(muestras//2 + (skip_factor-1)//2)
  picos_arreglados= picos+retraso

  #PLOT
  if plotting:

    inicio = 0
    final = inicio + segundos*fs

    plt.figure(figsize=(15,5))
    plt.plot(t[inicio:final],x[inicio:final])
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.title('Señal original')
    plt.show()

    [z, p, k] = scipy.signal.butter(orden_N, wn, btype='bandpass', fs=fs, output='zpk')
    [w, h] = scipy.signal.freqz_zpk(z, p, k, fs=1000)

    fig = plt.figure(figsize=(15,5))
    ax1 = fig.add_subplot(1, 1, 1)
    ax1.set_title('Diagrama de Bode del filtro pasabanda')
    ax1.semilogx(w, 20 * np.log10(abs(h)), 'b')
    ax1.set_ylabel('Ganancia [dB]')
    ax1.set_xlabel('Frecuencia [Hz]')
    ax1.grid()
    plt.xlim([1, 300])
    plt.ylim([-125, 10])
    plt.axvline(5, color='green') # cutoff frequency
    plt.axvline(15, color='green') # cutoff frequency
    plt.show()

    plt.figure(figsize=(15,5))
    plt.plot(t[inicio:final],x[inicio:final], label='Señal original')
    plt.plot(t[inicio:final],xfiltrada[inicio:final],'r', label='Señal filtrada')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.title('Señal filtrada con filtro pasabanda')
    plt.legend(loc='upper right')
    plt.show()

    plt.figure(figsize=(15,5))
    plt.plot(t,x_deriv,'r')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.title('Señal derivada')
    plt.show()

    plt.figure(figsize=(15,5))
    plt.plot(t,x_cuad,'r')
    plt.title('Señal elevada al cuadrado')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.show()

    plt.figure(figsize=(15,5))
    plt.plot(t,x_integ)
    plt.title('Señal integrada')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.show()

    plt.figure(figsize=(15,5))
    plt.plot(t,x_integ)
    plt.plot(t[picos],x_integ[picos],'ro')
    plt.axhline(y=umbral, color='g')
    plt.title('Señal integrada con los picos')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.show()

    x=np.array(x)
    plt.figure(figsize=(15,5))
    plt.plot(t,x)
    plt.plot(t[picos_arreglados],x[picos_arreglados],'ro')
    plt.title('Señal original con los picos')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.show()

  return picos_arreglados
```

A modo de ejemplo se hace la detección de complejos QRS en la derivación V5 del primer paciente (Camila Aristegui).

```python
marcas = []
flag = 1
for i in range(0, len(diccionario['ECG'])):
  for j in range(0, len(diccionario['fiduciario'][0][0][0])):
    if len(diccionario['fiduciario'][i][0][0][j]) < 10:
      flag = 0
  
  if flag:
    marcas.append(detector_QRS(diccionario['ECG'][i][:,4], fs, plotting = False))
  else:
    marcas.append([-1])
  flag = 1

print(len(marcas))
```

Para calcular la amplitud pico a pico del complejo QRS se utiliza la diferencia entre el valor máximo y mínimo del intervalo comprendido entre QRSon y QRSend (puntos fiduciales). Luego, se realiza el promedio ordinario para obtener una amplitud representativa del complejo QRS para toda la derivación V5 de cada paciente.

```python
# 0:R ; 1:T ; 2:QRSon ; 3:QRSend ; 4:Ton ; 5:Tend

ampQRS = []
for i in range(0, len(diccionario['ECG'])):
  k = 0
  suma = 0
  if len(senales_filtradas[i]) != 1:
    for j in range(0, min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][3]))):
      QRSon = diccionario['fiduciario'][i][0][0][2][j][0]
      QRSend = diccionario['fiduciario'][i][0][0][3][j][0]
      if m.isnan(QRSon) == 0 and m.isnan(QRSend) == 0 and QRSon < QRSend:
        QRSon = int(QRSon)
        QRSend = int(QRSend)
        dif = max(senales_filtradas[i][QRSon:QRSend]) - min(senales_filtradas[i][QRSon:QRSend])
        suma += dif
        k += 1
    promedio = suma/k
    a = min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][3]))
    promedio = suma + (a-k)*promedio
    promedio = promedio/a
    ampQRS.append(promedio)
  else:
    ampQRS.append(-1)
print(ampQRS)
```

### Pendiente de subida del complejo QR

Para calcular la pendiente de subida del complejo QR se utiliza la siguiente fórmula:

$\frac{Y_R - Y_{QRSon}}{(R - QRSon)/fs}$

donde Y es la amplitud de la señal en los puntos R y QRSon (puntos fiduciales) y fs la frecuencia de muestreo.

Luego, se realiza el promedio ordinario para obtener una pendiente de subida representativa del complejo QR para toda la derivación V5 de cada paciente.

```python
# 0:R ; 1:T ; 2:QRSon ; 3:QRSend ; 4:Ton ; 5:Tend

upQRS = []
for i in range(0, len(diccionario['ECG'])):
  k = 0
  suma = 0
  if len(senales_filtradas[i]) != 1:
    for j in range(0, min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][0]))):
      QRSon = diccionario['fiduciario'][i][0][0][2][j][0]
      R = diccionario['fiduciario'][i][0][0][0][j][0]
      if m.isnan(QRSon) == 0 and m.isnan(R) == 0 and QRSon < R:
        QRSon = int(QRSon)
        R = int(R)
        pendiente = (senales_filtradas[i][R] - senales_filtradas[i][QRSon])/((R - QRSon)/fs)
        suma += pendiente
        k += 1
    promedio = suma/k
    a = min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][3]))
    promedio = suma + (a-k)*promedio
    promedio = promedio/a
    upQRS.append(promedio)
  else:
    upQRS.append(-1)

print(upQRS)
```

### Pendiente de bajada del complejo RS

Para calcular la pendiente de bajada del complejo RS se utiliza la siguiente fórmula:

$\frac{Y_{QRSend} - Y_{R}}{(QRSend - R)/fs}$

donde Y es la amplitud de la señal en los puntos QRSend (punto fiducial) y R (máximo entre QRSon y QRSend), y fs la frecuencia de muestreo.

Luego, se realiza el promedio ordinario para obtener una pendiente de bajada representativa del complejo RS para toda la derivación V5 de cada paciente.

```python
# 0:R ; 1:T ; 2:QRSon ; 3:QRSend ; 4:Ton ; 5:Tend

downQRS = []
for i in range(0, len(diccionario['ECG'])):
  k = 0
  suma = 0
  if len(senales_filtradas[i]) != 1:
    for j in range(0, min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][3]))):
      QRSon = diccionario['fiduciario'][i][0][0][2][j][0]
      QRSend = diccionario['fiduciario'][i][0][0][3][j][0]
      if m.isnan(QRSon) == 0 and m.isnan(QRSend) == 0 and QRSon < QRSend:
        QRSon = int(QRSon)
        QRSend = int(QRSend)
        R = int(np.where(senales_filtradas[i] == max(senales_filtradas[i][QRSon:QRSend]))[0])
        pendiente = (senales_filtradas[i][QRSend] - senales_filtradas[i][R])/((QRSend - R)/fs)
        suma += pendiente
        k += 1
    promedio = suma/k
    a = min(len(diccionario['fiduciario'][i][0][0][2]), len(diccionario['fiduciario'][i][0][0][3]))
    promedio = suma + (a-k)*promedio
    promedio = promedio/a
    downQRS.append(promedio)
  else:
    downQRS.append(-1)

print(downQRS)
```

### Pendiente de subida de la onda T

Para calcular la pendiente de subida de la onda T se utiliza la siguiente fórmula:

$\frac{Y_{T} - Y_{Ton}}{(T - Ton)/fs}$

donde Y es la amplitud de la señal en los puntos Ton y T (puntos fiduciales), y fs la frecuencia de muestreo.

Luego, se realiza el promedio ordinario para obtener una pendiente de subida representativa de la onda T para toda la derivación V5 de cada paciente.

```python
# 0:R ; 1:T ; 2:QRSon ; 3:QRSend ; 4:Ton ; 5:Tend

upT = []
for i in range(0, len(diccionario['ECG'])):
  k = 0
  suma = 0
  if len(senales_filtradas[i]) != 1:
    for j in range(0, min(len(diccionario['fiduciario'][i][0][0][4]), len(diccionario['fiduciario'][i][0][0][1]))):
      Ton = diccionario['fiduciario'][i][0][0][4][j][0]
      T = diccionario['fiduciario'][i][0][0][1][j][0]
      if m.isnan(Ton) == 0 and m.isnan(T) == 0 and Ton < T:
        Ton = int(Ton)
        T = int(T)
        pendiente = (senales_filtradas[i][T] - senales_filtradas[i][Ton])/((T - Ton)/fs)
        suma += pendiente
        k += 1
    promedio = suma/k
    a = min(len(diccionario['fiduciario'][i][0][0][4]), len(diccionario['fiduciario'][i][0][0][5]))
    promedio = suma + (a-k)*promedio
    promedio = promedio/a
    upT.append(promedio)
  else:
    upT.append(-1)

print(upT)
```

### Pendiente de bajada de la onda T

Para calcular la pendiente de bajada de la onda T se utiliza la siguiente fórmula:

$\frac{Y_{Tend} - Y_{T}}{(Tend - T)/fs}$

donde Y es la amplitud de la señal en los puntos Ton (punto fiducial) y T (máximo del intervalo comprendido entre los puntos fiduciales Ton y Tend), y fs la frecuencia de muestreo.

Luego, se realiza el promedio ordinario para obtener una pendiente de bajada representativa de la onda T para toda la derivación V5 de cada paciente.

```python
# 0:R ; 1:T ; 2:QRSon ; 3:QRSend ; 4:Ton ; 5:Tend

downT = []
for i in range(0, len(diccionario['ECG'])):
  k = 0
  suma = 0
  if len(senales_filtradas[i]) != 1:
    for j in range(0, min(len(diccionario['fiduciario'][i][0][0][4]), len(diccionario['fiduciario'][i][0][0][5]))):
      Ton = diccionario['fiduciario'][i][0][0][4][j][0]
      Tend = diccionario['fiduciario'][i][0][0][5][j][0]
      if m.isnan(Ton) == 0 and m.isnan(Tend) == 0 and Ton < Tend:
        Ton = int(Ton)
        Tend = int(Tend)
        T = int(np.where(senales_filtradas[i] == max(senales_filtradas[i][Ton:Tend]))[0])
        pendiente = (senales_filtradas[i][Tend] - senales_filtradas[i][T])/((Tend - T)/fs)
        suma += pendiente
        k += 1
    promedio = suma/k
    a = min(len(diccionario['fiduciario'][i][0][0][4]), len(diccionario['fiduciario'][i][0][0][5]))
    promedio = suma + (a-k)*promedio
    promedio = promedio/a
    downT.append(promedio)
  else:
    downT.append(-1)

print(downT)
```

## Análisis estadístico de las características elegidas

Debido a que no sabemos si estos datos presentan una distribución normal, no podemos usar la prueba T de Student para obtener el valor P y hacer el análisis estadístico correspondiente. Como alternativa se utiliza el Wilcoxon Rank-Sum Test, que asume que los dos conjuntos de datos analizados presentan la misma distribución estadística.

Para esto usamos la función **ranksums()** de la libreria scipy.stats, lo cual nos devuelve el valor P de la prueba para cada característica. Luego, se definió como criterio de significancia estadística un valor P menor al 5% (P < 0.05).

**Entrenamiento**

Para un entrenamiento más efectivo se construye una función (**scramble_data**) que desordena aleatoriamente el conjunto de datos de entrenamiento y se realiza el entrenamiento 1000 veces con los datos en distinto orden. Para el entrenamiento se utiliza el método **.fit()**.

```python
def scramble_data(x, y):
  L = len(x)
  posiciones = list(range(0, L))
  x_scramble = np.zeros([L, 3])
  y_scramble = np.zeros(L)

  for i in range(0, L):
    num = int(random.choice(posiciones))
    posiciones.pop(posiciones.index(num))

    x_scramble[num] = x[i,:]
    y_scramble[num] = y[i]
  
  return x_scramble, y_scramble
```
